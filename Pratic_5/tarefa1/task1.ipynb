```python
#!apt-get update
#!apt-get install openjdk-8-jdk-headless
#!wget -c https://dlcdn.apache.org/spark/spark-3.3.1/spark-3.3.1-bin-hadoop3.tgz
#!tar -xzf spark-3.3.1-bin-hadoop3.tgz
#!pip install -q findspark

import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"
os.environ["SPARK_HOME"] = "/content/spark-3.3.1-bin-hadoop3"

import findspark
findspark.init()

from pyspark.sql import SparkSession
sc = SparkSession.builder.master("local[1]") \
.appName('SimpleApp') \
.getOrCreate() \
#.sparkContext

#Carregar o arquivo conjunto2.csv no colab
df = sc.read.options(header=True, delimiter=';').csv("conjunto2.csv")

#Transformar em rdd o df
rdd1 = df.rdd

#Mapear os campos de cargo e nome, fazendo o cargo como key e nome como value
rdd2 =rdd1.map(lambda x: (x["CARGO"], x["NOME"]))

#Teste para verificar os elementos
#for element in rdd2.collect():
#    print(element)

#Retirar todas as duplicatas
rdd3 = rdd2.distinct()

#Teste para verificar os elementos
#for element in rdd3.collect():
#    print(element)

#contagem dos elementos pela key, no caso o cargo
#assim, retornando na tela apenas os cargos com as quantidades
rdd3.countByKey()
```
